```python
# ! pip install requests
# ! pip install bs4

#Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
#and make data frame.


import requests
import pandas as pd
from bs4 import BeautifulSoup

url="https://www.imdb.com/chart/top/?ref_=nv_mv_250"
page=requests.get(url)
page

soup=BeautifulSoup(page.content,'html.parser')


content=soup.find('tbody',{'class':'lister-list'}).find_all('tr')
data=[]
i=1
for link in content:
    if i<101:
        name=link.find('td',{'class':'titleColumn'}).a.text
        year=str(link.find('span',{'class':'secondaryInfo'}).text.strip('()'))
        rank=str(link.find('td',{'class':'titleColumn'}).get_text(strip=True).split('.')[0])
        rating=str(link.find('td',{'class':'ratingColumn imdbRating'}).get_text(strip=True))
        data.append([name,year,rank,rating])
        i=i+1

imdb=pd.DataFrame(data,columns=['MovieName','year','rank','rating'])  
imdb


    
    

```

```python

#Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of
#release and make data frame


import requests
import pandas as pd
from bs4 import BeautifulSoup

url="https://www.imdb.com/india/top-rated-indian-movies/"
page=requests.get(url)
page

soup=BeautifulSoup(page.content,'html.parser')


indmovies=soup.find('tbody',{'class':'lister-list'}).find_all('tr')
data=[]
i=1
for link in indmovies:
    if i<101:
        moviename=link.find('td',{'class':'titleColumn'}).a.text
        movieyear=str(link.find('span',{'class':'secondaryInfo'}).text.strip('()'))
        movierank=str(link.find('td',{'class':'titleColumn'}).get_text(strip=True).split('.')[0])
        movierating=str(link.find('td',{'class':'ratingColumn imdbRating'}).get_text(strip=True))
        data.append([moviename,movieyear,movierank,movierating])
        i=i+1

imdb=pd.DataFrame(data,columns=['MovieName','MovieYear','MovieRank','MovieRating'])
imdb
```

```python
#) Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp . 


#! pip install requests
#! pip install bs4

import requests
import pandas as pd
from bs4 import BeautifulSoup

url='https://meesho.com/bags-ladies/pl/p7vbp'

soup=requests.get(url)

page=BeautifulSoup(soup.content,'html.parser')

link1=page.find_all('p',class_='Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS')

product=[]
for i in link1:
    product.append(i.text)

price=[]
link2=page.find_all('h5',class_='Text__StyledText-sc-oo0kvp-0 hiHdyy')
for i in link2:
    price.append(i.text)

discount=[]
link3=page.find_all('p',class_='Text__StyledText-sc-oo0kvp-0 fCJVtz NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm')
for i in link3:
    discount.append(i.text)

data=pd.DataFrame({'Product':product,
                  'Price':price,
                  'Discount':discount})
data
```

```python
# Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
#a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.




import requests
import pandas as pd
from bs4 import BeautifulSoup


url='https://www.icc-cricket.com/rankings/mens/team-rankings/odi'

soup=requests.get(url)

page=BeautifulSoup(soup.content,'html.parser')

TeamOneName=page.find('td',class_='rankings-block__banner--team-name').text.strip().replace('\nNZ','')
#TeamOneName=TeamOne.find('span',class_='u-hide-phablet')
TeamOneMatches=page.find('td',class_='rankings-block__banner--matches').text
TeamOneRanking=page.find('td',class_='rankings-block__banner--pos').text
TeamOneRating=page.find('td',class_='rankings-block__banner--rating u-text-right').text.strip()

line=page.find_all('tr',class_='table-body')

data=[]
data.append([TeamOneRanking,TeamOneName,TeamOneMatches,TeamOneRating])

for i in line:
    TeamRanking=i.find('td',class_='table-body__cell table-body__cell--position u-text-right').text.strip()
    Name=i.find('td',class_='table-body__cell rankings-table__team')
    TeamName=Name.find('span',class_='u-hide-phablet').text
    TeamMatches=i.find('td',class_='table-body__cell u-center-text').text
    TeamRating=i.find('td',class_='table-body__cell u-text-right rating').get_text(strip=True)
    
    data.append([TeamRanking,TeamName,TeamMatches,TeamRating])
name =pd.DataFrame(data,columns=['TeamRanking','TeamName','TeamMatches','TeamRating'])
name


```

```python

# 9) Write a python program to scrape mentioned details from dineout.co.in :
# i) Restaurant name
# ii) Cuisine
# iii) Location
# iv) Ratings
# v) Image URL


import requests
import pandas as pd
from bs4 import BeautifulSoup

url='https://www.dineout.co.in/delhi-restaurants/buffet-special'
soup=requests.get(url)

page=BeautifulSoup(soup.content,'html.parser')

RestName=[]
Location=[]
Ratings=[]
Images=[]


for i in page.find_all('div',class_='restnt-info cursor'):
    RestName.append(i.a.text)

    
for i in page.find_all('div',class_='restnt-loc ellipsis'):
    Location.append(i.text)

    
for i in page.find_all('div',class_='restnt-rating rating-4'):
    Ratings.append(i.text)
    
for i in page.find_all('img',class_='no-img'):
    Images.append(i['data-src'])
    
rest=pd.DataFrame({'RestName':RestName,'Location':Location,'Ratings':Ratings})
rest
    
```

```python

# 10) Write a python program to scrape first 10 product details which include product name , price , Image URL from
# https://www.bewakoof.com/women-tshirts?ga_q=tshirts .


# 10) Write a python program to scrape first 10 product details which include product name , price , Image URL from
# https://www.bewakoof.com/women-tshirts?ga_q=tshirts .


import requests
import pandas as pd
from bs4 import BeautifulSoup

page = requests.get("https://www.bewakoof.com/women-tshirts?ga_q=tshirts")
# page content
soup=BeautifulSoup(page.content)

# scraping the product name
product_name = [] # creating empty list
for i in soup.find_all('div',class_="productCardDetail")[:10]: 
    product_name.append(i.text)

# scraping the product price   
price = []
for i in soup.find_all('span',class_="discountedPriceText")[:10]: 
    price.append(i.text)
    
# scraping the image_url 
image_url = []
for i in soup.find_all('img',class_="productImgTag")[:10]: 
    image_url.append(i.get('src'))  
    

# Creating data frame 
import pandas as pd
df = pd.DataFrame({'Product_Name':product_name,'Price':price,'Image_url':image_url})
df
```
