```python
# Q.1) Scrape the details of most viewed videos on YouTube from Wikipedia.
# Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
# You need to find following details:
# A) Rank
# B) Name
# C) Artist
# D) Upload date
# E) Views


from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'

driver=webdriver.Chrome('chromedriver.exe')
driver.get(url)
driver.maximize_window()

df= pd.DataFrame(columns=['Rank', 'Name', 'Artist', 'Views', 'Date'])

Rank = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][0]/tbody/tr/td[1]')
Name = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][0]/tbody/tr/td[2]')
Artist = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][0]/tbody/tr/td[3]')
Views = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][0]/tbody/tr/td[4]')
Date = driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][0]/tbody/tr/td[5]')
```

```python
for i in range(len(Rank)):
    df= df.append({
        'Rank': Rank[i].text,
        'Name': Name[i].text,
        'Artist': Artist[i].text,
        'Views':Views[i].text,
        'Date': Date[i].text
    },ignore_index=True)
```

```python
#Q.8) Scrape the details of Highest selling novels.
#Url = https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-greycompare/
#You have to find the following details:
#A) Book name
#B) Author name
#C) Volumes sold
#D) Publisher
#E) Genre


from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

url = 'https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'

driver=webdriver.Chrome('chromedriver.exe')
driver.get(url)
driver.maximize_window()

df= pd.DataFrame(columns=['Book_name', 'Author_name', 'Volumes', 'Publisher','Genre'])

Book_name = driver.find_elements_by_xpath('//table[@class="in-article sortable"]/tbody/tr/td[1]')
Author_name = driver.find_elements_by_xpath('//table[@class="in-article sortable"]/tbody/tr/td[2]')
Volumes = driver.find_elements_by_xpath('//table[@class="in-article sortable"]/tbody/tr/td[3]')
Publisher = driver.find_elements_by_xpath('//table[@class="in-article sortable"]/tbody/tr/td[4]')
Genre = driver.find_elements_by_xpath('//table[@class="in-article sortable"]/tbody/tr/td[5]')
```

```python
for i in range(len(Book_name)):
    df= df.append({
        'Book_name': Book_name[i].text,
        'Author_name': Author_name[i].text,
        'Volumes': Volumes[i].text,
        'Publisher':Publisher[i].text,
        'Genre': Genre[i].text
    },ignore_index=True)
```

```python
# Q.9)  Scrape the details most watched tv series of all time from imdb.com.
# Url = https://www.imdb.com/list/ls095964455/
# You have to find the following details:
# A) Name
# B) Year span
# C) Genre
# D) Run time
# E) Ratings
# F) Votes

import requests
import pandas as pd
from bs4 import BeautifulSoup

url="https://www.imdb.com/list/ls095964455/"
page=requests.get(url)
page

soup=BeautifulSoup(page.content,'html.parser')


content=soup.find('div',{'class':'lister-list'}).find_all('div',{'class':'lister-item mode-detail'})
data=[]
i=1
for link in content:
    if i<51:
        name=link.find('h3',{'class':'lister-item-header'}).a.text
        year=str(link.find('span',{'class':'lister-item-year text-muted unbold'}).text.strip('()'))
        rank=str(link.find('span',{'class':'lister-item-index unbold text-primary'}).text.strip('()'))
        runtime=str(link.find('span',{'class':'runtime'}).text.strip('()'))
        genre=str(link.find('span',{'class':'genre'}).text.strip('()'))
        rating=str(link.find('span',{'class':'ipl-rating-star__rating'}).get_text(strip=True))
        data.append([rank,name,year,genre,runtime,rating])
        i=i+1

imdb=pd.DataFrame(data,columns=['Rank','SeriesName','Year','Genre','Runtime','Rating'])  
imdb
```
