```python
# Q.6). Write a program to scrap details of all the funding deals for second quarter (i.e Jan 21 – March 21) from trak.in.


import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# GETTING URL

url='https://trak.in/india-startup-funding-investment-2015/'

soup=requests.get(url)

page=BeautifulSoup(soup.content,'html.parser')
```

```python
funding=[]
# scrapping 2nd row of the page

fundings=page.find_all('tr',class_='row-2 even')
```

```python
data=[]

# scrapping startup name,investor,date,amount etc from 2nd row

for i in fundings:
    startup_name=(i.find('td',class_='column-3').text)
    date=(i.find('td',class_='column-2').text)
    industry=(i.find('td',class_='column-4').text)
    loc=(i.find('td',class_='column-6').text)
    investor=(i.find('td',class_='column-7').text)
    amount=(i.find('td',class_='column-9').text)
    data.append([date,startup_name,industry,loc,investor,amount])
    
```

```python
# Q.10) Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” 
# location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, 
# privates from price, dorms from price, facilities and property description.

# Connecting to web driver

driver=webdriver.Chrome('chromedriver.exe')
```

```python
url="https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2022-01-28&to=2022-01-31&guests=2&page=1"
driver.get(url)
```

```python
# Extracting hostel names 

hostel_names=driver.find_elements_by_xpath('//h2[@class="title title-6"]')

hostel=[]
for i in hostel_names:
    hostel.append(i.text)
```

```python
# Extracting distance from city center

distances=driver.find_elements_by_xpath('//div[@class="subtitle body-3"]')

distance=[]
for i in distances:
    distance.append(i.text)
```

```python
# Extracting  ratings

ratings=driver.find_elements_by_xpath('//div[@class="score orange big"]')

rating=[]
for i in ratings:
    rating.append(i.text)
```

```python
# Extracting  review

reviews=driver.find_elements_by_xpath('//div[@class="keyword"]')

review=[]
for i in reviews:
    review.append(i.text)
```

```python
# Extracting  total reviews

total_reviews=driver.find_elements_by_xpath('//div[@class="reviews"]')

tr=[]
for i in total_reviews:
    tr.append(i.text)
```

```python
# Extracting  privates from price

privates_price=driver.find_elements_by_xpath('//div[@class="price title-5"]')

private_price=[]
for i in privates_price:
    private_price.append(i.text)
```

```python
# Displaying extracted information for hostel names,ratings,distance,review,total review,privates from price

hostels=pd.DataFrame({'Hostel_Names':hostel[0:25],'Distance':distance[0:25],'Rating':rating[0:25],'Review':review[0:25],'Total_Reviews':tr[0:25],'PrivatesFromPrice':private_price[0:25]})
hostels
```
